Q7. Affichez le nombre d’organismes pour chaque royaume.

````python
#getting each word from each line
file = open("overview.txt", "r")
lines = file.readlines()

#linesArr will be a list of lists where the lists are an array of each words present in a line 
linesArr = []
for x in lines:
    linesArr.append(x.split("\t"))
#linesArr
````

````python
#getting the number of organisms for each kingdom 
nbBacteria = 0
nbViruses = 0
nbEukaryota = 0
nbArchaea = 0

for x in linesArr:
    match x[1]:
        case "Bacteria":
            nbBacteria += 1
        case "Viruses":
            nbViruses += 1
        case "Eukaryota":
            nbEukaryota += 1
        case "Archaea":
            nbArchaea += 1
        
print("The number of organisms in the Bacteria Kingdom is :",nbBacteria)
print("The number of organisms in the Viruses Kingdom is :",nbViruses)
print("The number of organisms in the Eukaryota Kingdom is :",nbEukaryota)
print("The number of organisms in the Archaea Kingdom is :",nbArchaea)

nbBacteria+nbViruses+nbEukaryota+nbArchaea
````

### Groupes

Q8. Qu’est-ce qu’un groupe ?

```markdown
Dans cette database, les groupes représentent chacun un des sous-groupes du Kingdom Group auquel l'organisme appartient. Ils correspondent à ce qu'on appelle le phylum et regroupent les organismes selon leurs caractéristiques communes.
```

Q9. Combien de groupes sont présents par royaume ?

````python
archaeaGroups = []
bacteriaGroups = []
eukaryotaGroups = []
virusesGroups = []

for x in linesArr:
    match x[1]:
        case "Bacteria":
            if x[2] not in bacteriaGroups:
                bacteriaGroups.append(x[2])
            else:
                continue
        case "Viruses":
            if x[2] not in virusesGroups:
                virusesGroups.append(x[2])
            else:
                continue
        case "Eukaryota":
            if x[2] not in eukaryotaGroups:
                eukaryotaGroups.append(x[2])
            else:
                continue
        case "Archaea":
            if x[2] not in archaeaGroups:
                archaeaGroups.append(x[2])
            else:
                continue

print("The number of groups in the Bacteria Kingdom is :",len(bacteriaGroups))
print("The number of groups in the Viruses Kingdom is :",len(virusesGroups))
print("The number of groups in the Eukaryota Kingdom is :",len(eukaryotaGroups))
print("The number of groups in the Archaea Kingdom is :",len(archaeaGroups))
````
